```python
from collections import deque


class NodoG:
    def __init__(self, estado):
        self.estado = estado
        self.granjero, self.lobo, self.cabra, self.col = estado

    def __str__(self):
        return f"Granjero: [{self.granjero}], Lobo: [{self.lobo}], Cabra: [{self.cabra}], Col: [{self.col}]"

    def get_sucesores(self):
        sucesores = []
        if self.granjero == 'i':
            nuevo_estado = ('d', self.lobo, self.cabra, self.col)
            if validar_estado(nuevo_estado):
                sucesores.append(NodoG(nuevo_estado))
            nuevo_estado = ('d', 'd',self.cabra, self.col)
            if validar_estado(nuevo_estado):
                sucesores.append(NodoG(nuevo_estado))
            nuevo_estado = ('d', self.lobo, 'd', self.col)
            if validar_estado(nuevo_estado):
                sucesores.append(NodoG(nuevo_estado))
            nuevo_estado = ('d', self.lobo, self.cabra, 'd')
            if validar_estado(nuevo_estado):
                sucesores.append(NodoG(nuevo_estado))
        else:
            # si el granjero puede cruzar solo
            nuevo_estado = ('i', self.lobo, self.cabra, self.col)
            if validar_estado(nuevo_estado):
                sucesores.append(NodoG(nuevo_estado))
            nuevo_estado = ('i', 'i', self.cabra, self.col)
            if validar_estado(nuevo_estado):
                sucesores.append(NodoG(nuevo_estado))
            nuevo_estado = ('i', self.lobo, 'i', self.col)
            if validar_estado(nuevo_estado):
                sucesores.append(NodoG(nuevo_estado))
            nuevo_estado = ('i', self.lobo, self.cabra, 'i')
            if validar_estado(nuevo_estado):
                sucesores.append(NodoG(nuevo_estado))
        return sucesores

def validar_estado(estado):# Reglas
    # granjero lobo
    if estado[0] == estado[1]:
        return True
    # lobo y cabra pero lobo no es igual al granjero
    elif estado[1] == estado[2] and estado[0] != estado[1]:
        return False
    # cabra y col pero cabra not granjero
    elif estado[1] == estado[3] and estado[0] != estado[1]:
        return False
    else:
        return True

def bfs(n_inicial, n_objetivo):
    visitados = set()
    cola = deque([(n_inicial, [])])
    while cola:
        (NodoG_actual, camino) = cola.popleft()
        if NodoG_actual.estado == n_objetivo.estado:
            return camino + [NodoG_actual]
        for sucesor in NodoG_actual.get_sucesores():
            if sucesor.estado not in visitados:
                visitados.add(sucesor.estado)
                cola.append((sucesor, camino+[NodoG_actual]))
    return None

def dfs(nodo_inicial, nodo_objetivo):
    visitados = set()
    pila = [(nodo_inicial, [])]

    while pila:
        nodo_actual, camino = pila.pop()

        if nodo_actual.estado == nodo_objetivo.estado:
            return camino + [nodo_actual]

        if nodo_actual.estado not in visitados and validar_estado(nodo_actual.estado):
            visitados.add(nodo_actual.estado)
            sucesores = nodo_actual.get_sucesores()

            for sucesor in sucesores:
                if sucesor.estado not in visitados:
                    pila.append((sucesor, camino + [nodo_actual]))

    return None

estado_inicial = NodoG(("i", "i", "i", "i"))
estado_objetivo = NodoG(("d", "d", "d", "d"))

solucion = bfs(estado_inicial,estado_objetivo)
if solucion is None:
    print('No se encontr贸 una soluci贸n al problema del granjero.')
else:
    for estado in solucion:
        print(estado)

print("DFS")
print()

profundidad = dfs(estado_inicial, estado_objetivo)

if profundidad is None:
    print('No se encontr贸 una soluci贸n al problema del granjero.')
else:
    for estado in profundidad:
        print(estado)
```
